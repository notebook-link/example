```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2Ï€
x = np.linspace(0, 2 * np.pi, 100)

# Compute sine of x
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue')
plt.title('Sine Function')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```
